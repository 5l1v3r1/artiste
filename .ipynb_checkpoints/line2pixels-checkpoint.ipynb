# Magnitude Regularization

```python
import numpy as np
import tensorflow as tf
from tensorflow.examples.tutorials.mnist import input_data
mnist = input_data.read_data_sets('MNIST_data', one_hot=True)

import time
import cPickle as pickle
import matplotlib.pyplot as plt
import matplotlib.cm as cm
%matplotlib inline
```

```python
batch = mnist.train.next_batch(1)
img = np.reshape(batch[0], (28,28))
plt.figure(0, figsize=(8,2))
plt.imshow(img, cmap=cm.gray)
```

```python
side = 28
img = np.zeros((side,side))
x = np.ones(12) ; y = -np.ones(12) #np.zeros(12)

px = np.cumsum(x) + side/2
py = np.cumsum(y) + side/2
for i in range(len(py)):
    if px[i] < side and px[i] >= 0 and px[i] < side and px[i] >= 0:
        img[py[i]-1:py[i]+1, px[i]-1:px[i]+1] = 1
plt.imshow(img)
```
