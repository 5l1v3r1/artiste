# mnist sketcher

```python
import numpy as np
import tensorflow as tf
from tensorflow.examples.tutorials.mnist import input_data
mnist = input_data.read_data_sets('MNIST_data', one_hot=True)

import time
import cPickle as pickle
import matplotlib.pyplot as plt
import matplotlib.cm as cm
%matplotlib inline
```

## Parameters

```python
batch_size = 32
global_step = 0
save_path = 'models/model.ckpt'
xlen = 28*28
ylen = 10
zlen = 2

tsteps = 30
reg_param = 1e-4
```

## Build vision model

```python
#awn
tf.reset_default_graph()
sess = tf.InteractiveSession()

x = tf.placeholder(tf.float32, shape=[None, xlen], name="x")
y_ = tf.placeholder(tf.float32, shape=[None, ylen], name="y_")
init = tf.truncated_normal_initializer(mean=0., stddev=.075, seed=1234, dtype=tf.float32)

reg = tf.contrib.layers.l2_regularizer(reg_param)
with tf.variable_scope("model", regularizer=reg):
    W = tf.get_variable("W", shape=[xlen,ylen], initializer=init)
y_hat = tf.nn.softmax(tf.matmul(x, W))
sess.run(tf.initialize_all_variables())
```

### Define variables

```python
model = {}
with tf.variable_scope('sketch',reuse=False):
    xavier_l1 = tf.truncated_normal_initializer(mean=0, stddev=1./np.sqrt(xlen), dtype=tf.float32)
    model['W1'] = tf.get_variable("W1", [xlen, 50], initializer=xavier_l1)

    rnn_init = tf.truncated_normal_initializer(stddev=0.075, dtype=tf.float32)
    model['rnn'] = tf.nn.rnn_cell.LSTMCell(50, state_is_tuple=True, initializer=rnn_init)

    model['state'] = model['rnn'].zero_state(batch_size=batch_size, dtype=tf.float32)

    xavier_l3 = tf.truncated_normal_initializer(stddev=1./np.sqrt(50), dtype=tf.float32)
    model['W3'] = tf.get_variable("W3", [50, zlen], initializer=xavier_l3)
```

### Initialize them

```python
def sketch_forward(x):
    state = model['state']

    h = tf.matmul(x, model['W1']) # tsteps list of [batches, H]
    h = tf.nn.relu(h)             # ReLU nonlinearity

    hs = [tf.squeeze(h_, [1]) for h_ in tf.split(1, tsteps, tf.reshape(h, [-1, tsteps, 50]))]
    rnn_outs, state = tf.nn.seq2seq.rnn_decoder(hs, state, model['rnn'])
    rnn_out = tf.reshape(tf.concat(1, rnn_outs), [-1, 50])
    
    logps = tf.matmul(rnn_out, model['W3'])
    p = tf.reshape(logps, [-1, zlen])
    return p
```

```python
yloss = tf.nn.l2_loss( y_ - y_hat ) / (ylen*batch_size) + sum(tf.get_collection(tf.GraphKeys.REGULARIZATION_LOSSES))
optimizer = tf.train.AdamOptimizer(1e-4)
grads = optimizer.compute_gradients(yloss, var_list=tf.trainable_variables())
train_op = optimizer.apply_gradients(grads)

zloss = (1 - tf.argmax(y_hat,1)) / batch_size
optimizer = tf.train.AdamOptimizer(1e-4)
grads = optimizer.compute_gradients(zloss, var_list=model.values())
train_op = optimizer.apply_gradients(grads)

sess.run(tf.initialize_all_variables())
```

## Model overview
